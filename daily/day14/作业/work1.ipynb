```python
import random

import matplotlib.pyplot as plt

# 绘制（1，4） （0，5） （9，6）
# x = [1,0,9]   # 手动输入
# x = range(8)    # 用range产生0-7序列

x = [random.randint(0,20) for _ in range(8)] 
y = [9,7,6,5,4,2,1,3]
plt.plot(x,y)
plt.show()
```

```python
# plot 的参数
# plot ( x , y , color , alpha=透明度 , linesyle='线条' , linewidth=粗细 ,marker='拐点' ,label='一图多线标签')

x = range(8)    
y = [9,7,6,5,4,2,1,3]
plt.plot(x,y,color='yellow',alpha=0.9,linestyle='-',linewidth=3,marker='o')
plt.show()
```

```python
x = range(13)
y = [random.randint(0,45) for _ in range(13)]     

x_ticks_label = ['{}:00'.format(i) for i in x]      # 用for i 把x列表的数都format
y_ticks_label = ['{}度'.format(j) for j in y]      

# 解决{}度 乱码的问题  导入字体包，设置字体
from matplotlib import font_manager as font
my_font = font.FontProperties(fname='C:\\Windows\\Fonts\\msyh.ttc',size=8)

plt.xticks(x,x_ticks_label,fontproperties=my_font)
plt.yticks(y,y_ticks_label,fontproperties=my_font,rotation=10)

# 为x，y轴设置标签
plt.xlabel('时间',fontproperties=my_font)
plt.ylabel('温度',fontproperties=my_font)


plt.plot(x,y,color='blue',alpha=0.9,linestyle='-',linewidth=2)
plt.show()
```

```python
x1 = range(13)
y1 = [random.randint(0,20) for _ in range(13)]  
x2 = range(13)
y2 = [random.randint(0,20) for _ in range(13)]  

x_ticks_label = ['{}岁'.format(i) for i in x1]
y_ticks_label = ['{}KG'.format(j) for j in y1]
my_font = font.FontProperties(fname='C:\\Windows\\Fonts\\msyh.ttc',size=8)
plt.xticks(x1,x_ticks_label,fontproperties=my_font)
plt.yticks(y1,y_ticks_label,fontproperties=my_font,rotation=10)
plt.xlabel('年龄',fontproperties=my_font)
plt.ylabel('体重',fontproperties=my_font)

plt.plot(x1,y1,color='blue',alpha=0.9,linestyle='-',label='样本A')
plt.plot(x2,y2,color='green',alpha=0.9,linestyle='--',label='样本B')

# 添加线条图例
plt.legend( prop=my_font, loc='upper right' )
# 添加分割线
plt.grid(alpha=0.6)

plt.show()
```

```python
# 用 nmpy 生成一个 1-100 的数组 作为每个子图的横坐标标签
import numpy as np
x = np.arange(1,100)

# 切割成四宫格,每一个 ax 都可以看作一个plt 
fig,axes = plt.subplots(2,2)

ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[1,0]
ax4 = axes[1,1]

ax1.plot(x,x)
ax2.plot(x,x**2)
ax3.plot(x,np.exp(x))
ax4.plot(x,np.log(x))

plt.show()
```

```python
# 为坐标轴设置切片 ::n

y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17, 18, 21, 16, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 22,
     23]
x = range( 1, 32 )

x_ticks_label = ['{}岁'.format(i) for i in x]
y_ticks_label = ['{}KG'.format(j) for j in y]

my_font = font.FontProperties(fname='C:\\Windows\\Fonts\\msyh.ttc',size=8)
plt.xticks(x[::2],x_ticks_label[::2],fontproperties=my_font)
plt.yticks(y,y_ticks_label,fontproperties=my_font,rotation=10)

# 错误写法 ： plt.yticks(y[::2],y_ticks_label[::2],fontproperties=my_font,rotation=10)
# 因为 ， 切片要求 1、局部数组有序  2、数组里数据唯一 3、切片大小不能大于局部数据大小

plt.grid(alpha=0.6) 
plt.scatter( x ,y )
```

```python
import matplotlib
label_list = ["第一部分", "第二部分", "第三部分"]  # 各部分标签
size = [55, 35, 10]    # 各部分大小
color = ["red", "green", "blue"]   # 各部分颜色
explode = [0, 0.05, 0] # 各部分突出值 -- 离中心的距离

# 设置matplotlib的默认字体
matplotlib.rcParams['font.sans-serif']=['SimHei']

# 绘制饼状图
patches, l_text, p_text = plt.pie(size,explode=explode, colors=color, labels=label_list,
                                  labeldistance=1.1, autopct="%1.1f%%", startangle=90, pctdistance=0.6
                                  )

# 设置横轴和纵轴大小相等，这样饼才是圆的
plt.axis("equal")  

# 添加各部分的图例
plt.legend()

plt.show()
```
