# 用 nmpy 读数据 ， 然后用 matplotlib 展示

```python
import numpy as np
import matplotlib.pyplot as plt
us_file_path = "../youtube_video_data/US_video_data_numbers.csv"
uk_file_path = "../youtube_video_data/GB_video_data_numbers.csv"

# 引入文件US_video_data_numbers.csv，ndarray类型
t1 = np.loadtxt(us_file_path,delimiter=",",dtype="int")
t2 = np.loadtxt(us_file_path,delimiter=",",dtype="int")

# 选择 t1 的最后一列（评论）
t_us_comments = t1[:,-1]

# 选择小于 5000 的评论 ， 数据过滤 ， 舍弃一些不必要的
t_us_comments = t_us_comments[t_us_comments<=5000]

print(t_us_comments.max(),t_us_comments.min())

d = 50

bin_nums = (t_us_comments.max()-t_us_comments.min())//d

plt.hist(t_us_comments,bin_nums)

plt.show()
```

```python
# 引入文件GB_video_data_numbers.csv，ndarrary类型
t_uk = np.loadtxt(uk_file_path,delimiter=",",dtype="int")

t_uk_comment = t_uk[:,-1]
t_uk_like = t_uk[:,1]

# 选择小于500000的unlike
t_uk = t_uk[t_uk[:,1]<=500000]

plt.figure(figsize=(20,8),dpi=80)
plt.scatter(t_uk_like,t_uk_comment)

plt.show()
```

```python
# ndarray数组赋值
# 创建一个 ndarray 数组
original_array = np.array([1, 2, 3, 4, 5])

# 赋值操作
assigned_array = original_array

# 修改赋值后的数组
assigned_array[0] = 100

print("原始数组（赋值后）:", original_array)
print("赋值后的数组:", assigned_array)

original_array = np.array([1, 2, 3, 4, 5])
assigned_array = original_array.copy()
print("原始数组（复制后）:", original_array)
print("copy()后的数组:", assigned_array)
```
